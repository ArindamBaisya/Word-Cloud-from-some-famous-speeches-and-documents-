<h1> Word Cloud from some famous speeches and documents </h1>
<p> </p>
In this fun project we analyze some famous speeches and documents from the pages of history and draw Word Clouds and find out what they have to say. Will we find something meaningful?

First, we will import the required packages

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

<b>The first speech we consider is the famous Gettysburg Address by Abraham Lincoln. Photo credit: Wikipedia </b>

<img src="Abraham Lincoln.jpg">

```python
txtfile = open('GettysburgAddress.txt', 'r')
s1 = txtfile.read()
```

```python
wc = WordCloud(background_color="ivory")
wc.generate(s1)
# Display the word cloud.
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.figure(figsize=(20,10))
plt.show()
```

Unsurprisingly, in this short but historic speech, delivered during the solemn occasion of the dedication of the Soldiers' National Cemetery in Gettysburg, the top words are "dedication", "nation" and "greatness".!
